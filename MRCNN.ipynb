1.4 Modify the Mask R-CNN Model

In the model.py file, we modify the model to prevent issues related to TensorFlow 2.x’s eager execution mode. Specifically, we disable the metrics_tensors:

```python
with open('mrcnn/model.py') as f:
    model_file = f.read()

with open('mrcnn/model.py', 'w') as f:
    model_file = model_file.replace("self.keras_model = self.build(mode=mode, config=config)",
                                    "self.keras_model = self.build(mode=mode, config=config)\n        self.keras_model.metrics_tensors = []")
    f.write(model_file)
```

Step 5: Set Up the Configuration
Create a custom configuration for inference, which adjusts the batch size and image processing settings. This configuration is derived from the pre-defined COCO configuration.

To create the model- first need to create a configuration object. This object defines properties for both training and prediction, including the number of classes, the GPU count, and the learning rate.

There is already a configuration available for the MS-COCO dataset that only needs to be slightly modified for inference. 

```python
class InferenceConfig(coco.CocoConfig):
    # Set batch size to 1 since we'll be running inference on
    # one image at a time. Batch size = GPU_COUNT * IMAGES_PER_GPU
    GPU_COUNT = 1
    IMAGES_PER_GPU = 1

config = InferenceConfig()
config.display()
```

This configuration ensures that the model processes images one at a time, which is common for inference tasks.

Step 7: Set Up Image for Detection
Before running object detection, make sure to have an image available to test the model on. Load and process the image in the correct format (RGB) for Mask R-CNN.

```python
import os
import cv2
from mrcnn import visualize

# Path to your image
image_path1 = "/Users/madhumitham/Downloads/cars.png"

# Load the image as a NumPy array
image1 = cv2.imread(image_path1)  # This reads the image in BGR format (OpenCV default)
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)  # Convert to RGB, as Mask R-CNN expects

# Run detection
results1 = model.detect([image1], verbose=1)

# Visualize results
r = results1[0]
visualize.display_instances(image1, r['rois'], r['masks'], r['class_ids'], 
                            class_names, r['scores'])
```
